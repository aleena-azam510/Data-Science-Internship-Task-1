# Data Science Task 1
## EDA and Visualization of a Real-World Dataset

### Importing required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading Dataset

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
df.head()
```

### Checking null/empty values

```python
df.isnull().sum()
```

```python
df.isnull().sum()/len(df)*100
```

### Filling missing values

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df['Cabin'].fillna(df['Cabin'].mode()[0],inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

### Cheking and Dropping Duplicates

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

### Creating bar charts for categorical variables

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Pclass',data=df)
plt.title('Passenger Class')
plt.show()
```

### Plot histograms for numeric distributions

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

### Generating a correlation heatmap for numeric features

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Drop any remaining NaN values (optional)
numeric_df = numeric_df.dropna()

# Plot heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation - Titanic Dataset")
plt.show()
```
